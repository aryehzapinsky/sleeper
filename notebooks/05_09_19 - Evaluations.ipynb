```python
import numpy as np
import pandas as pd
import ast
```

```python
movements_df = pd.read_csv('movements.csv')
movements_df.head(2)
```

```python
movements_df['total'] = (movements_df['body'] + 
                         movements_df['arm_and_head'] +
                         movements_df['arm'] +
                         movements_df['head']
                        )
movements_df.head(20)
```

```python
record_df = pd.read_csv('/Volumes/NO NAME/sleeper/record.csv', index_col=0)
#record = record[['folder_name','reduced_std_frames','count_reduced_std_frames']]
record_df['reduced_std_frames'] = record_df['reduced_std_frames'].apply(lambda x: ast.literal_eval(x))
record_df.head(20)
```

```python
# Start record from 4th index to start with videos -- 2019-02-25
selected_record_df = (record_df.loc[4:19, ['folder_name', 
                                          'count_std_frames',
                                         'count_reduced_std_frames']]
                      .reset_index(drop=True))
selected_record_df.head(2)
```

```python
diff = pd.DataFrame(np.abs(selected_record_df['count_std_frames'] - movements_df['total']),
                   columns=['diff'])
percentage_diff = pd.DataFrame(np.abs(selected_record_df['count_std_frames'] 
                               - movements_df['total'])/ movements_df['total'], 
                               columns=['percentage_diff'])

comparison_df = (pd.concat([movements_df['total'],
                   selected_record_df['count_std_frames'],
                   diff,
                   percentage_diff],
                 axis=1)

                )
display(comparison_df)
print("MEAN: \n{}".format(comparison_df.mean()))
print("MEDIAN: \n{}".format(comparison_df.median()))
```

```python
diff = pd.DataFrame(np.abs(selected_record_df['count_reduced_std_frames'] 
                           - movements_df['body']),
                   columns=['diff'])
percentage_diff = pd.DataFrame(np.abs(selected_record_df['count_reduced_std_frames'] 
                               - movements_df['body'])/ movements_df['body'], 
                               columns=['percentage_diff'])

comparison_df = (pd.concat([movements_df['body'],
                   selected_record_df['count_reduced_std_frames'],
                   diff,
                   percentage_diff],
                 axis=1)

                )
display(comparison_df)
print("MEAN: \n{}".format(comparison_df.mean()))
print("MEDIAN: \n{}".format(comparison_df.median()))
```
